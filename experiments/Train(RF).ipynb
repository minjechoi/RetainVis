```python
from sklearn.ensemble import RandomForestClassifier as RF
import os
os.chdir('..')
import numpy as np
import pickle
```

```python
with open('data/preprocessed/2014_out_dict2.pckl','rb') as f:
    out_dict = pickle.load(f)
```

```python
jids = list(out_dict.keys())
import random
random.shuffle(jids)
```

```python
in_array = []
out_array = []
cnt =0
for t in jids[:1000000]:
    cnt+=1
    V = out_dict[t]
    if len(V)==0:
        continue
    else:
        tmp_arr = np.zeros(1400,dtype=int)
        for tup in V:
            _,_,lst,out = tup
            for l in lst:
                tmp_arr[l]+=1
    in_array.append(tmp_arr)
    out_array.append(out)
    if cnt%1000==0:
        print(cnt)
in_array = np.array(in_array,dtype=int)
out_array = np.array(out_array,dtype=int)
tr_in = in_array[:700000]
t_in = in_array[700000:1000000]
tr_out = out_array[:700000]
t_out = out_array[700000:1000000]
```

```python
clf = RF()
clf.fit(tr_in,tr_out)
```

```python
# get top-1 scores
pred = clf.predict(t_in)
top1_acc = np.array(pred==t_out,dtype=int).sum()/len(t_out)
print("Top-1 recall: %1.3f" %top1_acc)
```

```python
# get top-5 scores
probs = clf.predict_proba(t_in)
```

```python
N = 5
indices = np.argpartition(probs, -N, axis=1)[:,-N:]
corr = 0
for i in range(len(t_out)):
    if t_out[i] in indices[i]:
        corr+=1
top5_acc = corr*1.0/len(t_out)
print("Top-5 recall: %1.3f" %top5_acc)
```
